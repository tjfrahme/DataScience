```python
# Linear Regression - linear approximation of a causal relationship b/ 
# two or more variables

# Y (preidcted) - Dependent Variable
# X - Independent Variable(s)
# Y = f(x1 + x2 +...)
#

# y = B0 + B1x1 + e
# B0 is the constant
# x1 is the variable
# e is error of estimation - diff b/ actual and predicte

# Correlation vs. Regression 
# Correlation - relationship, movement together, p(x,y) = p(y,x), depicted
# by a single point
#
# Regression - one variable affect the other, cause/effect, depicted by line on 
# a graph
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

#overides matplotlib and sets seaborn
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)  # represents x0

# output of Ordinary Least Squares Method
# fit () applies a specific estimation techinque to obtain the fit of the model
results = sm.OLS(y,x).fit()  
```

```python
x.head()
```

```python
results
```

```python
results.summary()
```

```python
# const = constant b1
# SAT = b2

# std error show the accuracy of prediction for each variable.  Lower error
# means better estimate

# P>|t|... p-value < 0.05 means the variable is significant  

# sum of squares total(SST) - measures the total variability of the dataset

# sum of squares regression(SSR) - measures the explained variability by your 
# line how well your line fits the data

# sum of squares error(SSE/RSS) - measures the unexeplained variability by the
#  regression

# SST = SSR + SSE

# R^2 - SSR/SST, between 0 and 1.  regression explains the variability
# ~41% SAT explains GPA
```

```python
plt.scatter(x1, y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1, yhat, lw = 4, c = 'orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
