```python
# Mutiple Reqression 
# goal is to min SSE

# y = B0 + B1x1 + B2X2 .... + e

# Adjusted R^2
# R^2 - measures how much of the total variability is explained by our model
# Adjusted R^2 penalizes excessive use of variables
# Basis for comparing different Regression models
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Rand 1,2,3']]
```

```python
x1.head()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
# Adj R^2 down because additional variable is useless
# also P>|t| for "Rand 1,2,3" is greater then .05
```

```python
# F-Statistic: 
# F-Test:
# H0: B1 = B2 = B3... = 0
# H1:  at least one Bi <> 0  (not equal)
# If all betas are 0, then none of the x's matter, our model is invalid

#^ F-Stat is 27.76, and Prob(F-Stat) or Pvalue is .00000006
# lower the F-Stat the closer to a non-significant model
```

Regression Assumptions

1.  Linearity - relationship exisits between variables that forms a straight line.  If not linear, transform it so it is

2. No endogeneity - prohibits link between independant variables and error.  Omitted Variable Bias, occurs when missing a variable

3.  Normality and homoscedasticity - assume the error term is normally distributed. Homoscedacity means the error term should have equal variance.  Can try and use logs

4.  No Autocorrelation - errors are assumed to be uncorrelated. To detect use Durbin-Watson.  Values
fall b/ 0-4.  2 - no correlation, below 1 and above 3 is cause for alarm

5.  No multicollinearity - 2 or more variables have a high corelation.  To fix, drop one of the variable. Transform them into one variable (avg price)

Dummy Variables

For non-numerical fields, need to convert into categorical (1's, 0's)

```python
data = pd.read_csv('1.03. Dummies.csv')
```

```python
data.head(20)
```

```python
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No': 0})
```

```python
data.head(20)
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
```

```python
x.head()
```

```python
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
# GPA = .6439 + .0014 * SAT + .2226 * Dummy   --- Dummy (0 or 1)
```

```python
# Predicting values
```

```python
new_data = pd.DataFrame({'const': 1, 'SAT': [1700, 1670], 'Attendance': [0,1]})
new_data = new_data[['const', 'SAT', 'Attendance']]
new_data
```

```python
new_data.rename(index={0:'Bob', 1:"Alice"})
```

```python
prediction = results.predict(new_data)
```

```python
prediction
```

```python
predictionsdf = pd.DataFrame({'Predictions': prediction})
joined = new_data.join(predictionsdf)
joined.rename(index={0:'Bob', 1:'Alice'})
```
