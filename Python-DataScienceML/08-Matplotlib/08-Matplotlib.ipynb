```python
%matplotlib inline  
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
# plt.show()  include this if you are in something like a IDE
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented method
fig = plt.figure()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()  # 1) create an empty figure

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])     # 2) set axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)                         # 3) plot off of these axes
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 2)

#axes.plot(x,y)
plt.tight_layout()     # fixes overlapping graphs
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[0].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize = (3,2))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
#saving file as a type
fig.savefig('my_picture.png',dpi = 200)
```

```python
# Adding a Legend
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared')
ax.plot(x,x**3, label ='X Cubed')

ax.legend(loc=0)  #location code cooses best spot
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', linewidth=5, linestyle='--')

#ax.plot(x,y,color='orange', linewidth=5, alpha=0.5)  # alpha changes transparency 

#ax.plot(x,y,color='FF8C00')
```

```python
# adding in marker sizes
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', linewidth=2, linestyle='--', marker='o',
        markersize='10', markerfacecolor='blue')
```

# Control over Axis Appearance and Plot Range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', linewidth=2)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
