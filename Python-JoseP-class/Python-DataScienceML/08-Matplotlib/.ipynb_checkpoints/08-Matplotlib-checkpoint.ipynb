```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  # allows you to see the charts in jupyter notebook
                    # at end if not using jupyter notebook plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
# plt.show()  include this if you are in something like a IDE
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented method
fig = plt.figure()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()  # 1) create an empty figure

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])     # 2) set axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)                         # 3) plot off of these axes
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
