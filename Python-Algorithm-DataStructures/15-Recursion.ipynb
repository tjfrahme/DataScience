```python
# Recursion 

# 2 instances
# Recursion is used as a technique in chiwch a function makes one or more calls
# to itself

# When a data structure uses smaller instances of the exact same type of 
# data structure when it represents itself

```

```python
# Factorial example
# if n = 0 then n! = 1

# 4! = 4 * 3 * 2 * 1
# n! = n * (n-1)!

def factorial(n):
    
    # Base Case
    if n == 0:
        return 1
    
    
    else:
        return n * factorial(n-1)
    
```

```python
factorial(5)
```
